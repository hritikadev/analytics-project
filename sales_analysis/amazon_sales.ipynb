```python
# removing the Rupees sign in diounted and actual price columns
# removing the comma and changing data type to float
df['discounted_price'] = df['discounted_price'].str.replace('₹','')
df['discounted_price'] = df['discounted_price'].str.replace(',','').astype('float64')

df['actual_price'] = df['actual_price'].str.replace('₹','')
df['actual_price'] = df['actual_price'].str.replace(',','').astype('float64')
```

```python
# removing the percent symbol and changing percent to decimal values(dividing by 100)
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')/100
```

```python
# crating new dataframe with the selected columns (columns related to reviews are left)
df1 = df[['product_id', 'product_name', 'category', 'discounted_price', 'actual_price', 
          'discount_percentage', 'rating', 'rating_count']].copy()
```

```python
# removing white spaces in product_id
df1['product_id'].str.strip()
```

```python
# expanding the catgories by spliting about '|'
categories = df1['category'].str.split('|', expand=True)
categories
```

```python
# replacing number by name inthe column name
categories = categories.rename(columns={0:'category_1', 1:'category_2', 2:'category_3'})
```

```python
df1.head(3)
```

```python
# adding the categiry columns into the dataframe df1
df1['category1'] = categories['category_1']
df1['category2'] = categories['category_2']
df1['category3'] = categories['category_3']

df1.head(3)
```

```python
# unique value with their count in column category1
df1['category1'].value_counts()
```

```python
# refining the category names
df1['category1'] = df1['category1'].str.replace('&', ' & ')
df1['category1'] = df1['category1'].str.replace('OfficeProducts', 'Office Products')
df1['category1'] = df1['category1'].str.replace('MusicalInstruments', 'Musical Instruments')
df1['category1'] = df1['category1'].str.replace('HomeImprovement', 'Home Improvement')
```

```python
# unique value with their count in column category2
df1['category2'].value_counts()
```

```python
# refining the category names
df1['category2'] = df1['category2'].str.replace('&', ' & ')
df1['category2'] = df1['category2'].str.replace('HomeAppliances', 'Home Appliances ')
df1['category2'] = df1['category2'].str.replace('HomeTheater', 'Home Theater')
df1['category2'] = df1['category2'].str.replace('WearableTechnology', 'Wearable Technology')
df1['category2'] = df1['category2'].str.replace('AirQuality', 'Air Quality ')
df1['category2'] = df1['category2'].str.replace('NetworkingDevices', 'Networking Devices')
df1['category2'] = df1['category2'].str.replace('OfficePaperProducts', 'Office Paper Products')
df1['category2'] = df1['category2'].str.replace('ExternalDevices', 'External Devices')
df1['category2'] = df1['category2'].str.replace('DataStorage', 'Data Storage')
df1['category2'] = df1['category2'].str.replace('GeneralPurposeBatteries', 'General Purpose Batteries')
df1['category2'] = df1['category2'].str.replace('BatteryChargers', 'Battery Chargers')
df1['category2'] = df1['category2'].str.replace('CraftMaterials', 'Craft Materials ')
df1['category2'] = df1['category2'].str.replace('OfficeElectronics', 'Office Electronics')
df1['category2'] = df1['category2'].str.replace('PowerAccessories', 'Power Accessories ')
df1['category2'] = df1['category2'].str.replace('CarAccessories', 'Car Accessories')
df1['category2'] = df1['category2'].str.replace('HomeMedicalSupplies', 'Home Medical Supplies')
```

```python
df1.head(3)
```

```python
# # unique value with their count in column category3
df1['category3'].value_counts()
```

```python
# dropping category3 as it has too many values
df1 = df1.drop(['category3'], axis = 'columns')
```

```python
# creating rating category for diffrent values of 'rating'
rating_cat = []

for score in df1['rating']:
    if score < 2.0 : rating_cat.append('Poor')
    elif score < 3.0 : rating_cat.append('Below Average')
    elif score < 4.0 : rating_cat.append('Average')
    elif score < 5.0 : rating_cat.append('Above Average')
    elif score == 5.0 : rating_cat.append('Excellent')
```

```python
# creating a new column for rating category
df1['rating_cat'] = rating_cat

# changing the datatype to 'category'
df1['rating_cat'] = df1['rating_cat'].astype('category')

# reordering the categories
df1['rating_cat'] = df1['rating_cat'].cat.reorder_categories(['Below Average', 'Average', 
                                                                  'Above Average', 'Excellent'], ordered=True)
```

```python
# creating new column for differnce in actual and discounted prices
df1['price_diff'] = df1['actual_price'] - df1['discounted_price']
```

**Preparing a prediction model**

```python
# For a simple linear regression model taking 
#Independent variable X as actual price
#Dependent variable Y as discounted price

X = df1.iloc[:, 4].values.reshape(-1, 1)
y = df1.iloc[:, 3].values.reshape(-1, 1)
```

```python
# importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score, KFold
```

```python
# Splitting the dataset int train and test data 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

# fitting the model
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# calculating the R2 score 

y_pred = reg.predict(X_test)
print('R2 Score: ', r2_score(y_test, y_pred))
```

```python
# plotting a scatter plot and regression line
plt.scatter(X_train, y_train)
plt.plot(X_test, y_pred, color='red')
plt.xlabel('Actual Price', fontweight='bold')
plt.ylabel('Discounted Price', fontweight='bold')
plt.show()
```

```python
# calculationg the cross validation result

kf = KFold(n_splits=10, shuffle=True, random_state=21)
cv_results = cross_val_score(reg, X, y, cv=kf)
print('Cross Validation Mean: ', cv_results.mean())
```
